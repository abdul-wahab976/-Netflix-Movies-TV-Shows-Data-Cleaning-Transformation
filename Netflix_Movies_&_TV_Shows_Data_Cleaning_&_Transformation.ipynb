## a). MESSAY DATA

- 📌 Structural Issues in Netflix Dataset

- Duration column (mixed values)

- Movies → minutes (e.g., 90 min)

- TV Shows → seasons (e.g., 2 Seasons)

- Problem → ek column me do alag types ki information hai → ideally split into two: duration_minutes & num_seasons.

- Listed_in column (genres/categories)

- Multiple categories ek hi cell me comma-separated (e.g., Comedies, Dramas, Thrillers).

- Problem → 1 cell ≠ 1 value (violates database normalization).

Ideally split into multiple rows or separate genre table.

- Cast column

Multiple actors ek hi cell me (e.g., "Shah Rukh Khan, Kajol, Rani Mukerji").

Problem → many-to-many relationship should not be in one column.

- Country column

Sometimes contains multiple countries (e.g., "United States, India").

Problem → should be split into separate rows for each country.

- Date_added column

Stored as text (e.g., "September 24, 2021").

```python
# ---------- 1. date_added ----------
# strip spaces + convert to datetime safely
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), errors='coerce')

# jo NaT (null) bache unko release_year se fill karo
df['date_added'] = df.apply(
    lambda row: pd.to_datetime(f"01-01-{row['release_year']}")
                if pd.isnull(row['date_added']) else row['date_added'],
    axis=1
)

# ---------- 2. director ----------
# null ko "Unknown Director" se fill
df['director'] = df['director'].fillna("Unknown Director")

# ---------- 3. cast ----------
# null ko "No Cast Info" se fill
df['cast'] = df['cast'].fillna("No Cast Info")

# ---------- 4. country ----------
# null ko "Unknown Country" se fill
df['country'] = df['country'].fillna("Unknown Country")

# ---------- 5. rating ----------
# forward fill lagao (pichle row ka value uthao)
df['rating'] = df['rating'].fillna(method='ffill')

# ---------- 6. duration ----------
# null ko "Unknown Duration" se fill
df['duration'] = df['duration'].fillna("Unknown Duration")
```

```python
#Test:
df.info()
```

```python
# 1. date_added → datetime format me convert
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# 2. duration → split into two columns
df[['duration_int', 'duration_type']] = df['duration'].str.extract(r'(\d+)\s*(\w+)')

# duration_int ko numeric me convert karna (int)
df['duration_int'] = pd.to_numeric(df['duration_int'], errors='coerce')
```

```python
#DRop value for missing:
df = df.dropna(subset=['duration_int', 'duration_type'])
```
